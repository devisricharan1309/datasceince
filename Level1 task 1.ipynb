```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("Dataset .csv")
tar="Aggregate rating"
plt.figure(figsize=(8,5))
sns.boxplot(x=df[tar])
plt.title("BOX PLOT")
plt.xlabel("Aggregate Rating")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df[tar],bins=30,kde=True,color="blue")
plt.title("HISTOGRAM")
plt.xlabel("Aggregate Rating")
plt.ylabel("Frequency")
plt.show()
```
