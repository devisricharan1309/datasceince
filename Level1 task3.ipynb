```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("Dataset .csv")
from shapely.geometry import Point
import geopandas as gpd
from geopandas import GeoDataFrame
gdf=gpd.GeoDataFrame(df,geometry=gpd.points_from_xy(df.Longitude,df.Latitude))
world=gpd.read_file(gpd.datasets.get_path("naturalearth_lowres"))
gdf.plot(ax=world.plot("continent",legend=True,figsize=(18,15)),marker="o")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(y=df["City"],order=df["City"].value_counts().head(20).index)
plt.title("DISTRIBUTION OF RESTAURANTS ACROSS CITIES")
plt.xlabel("No.of Restaurants")
plt.ylabel("No.of cities")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("Dataset .csv")
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['Rating text'] = label_encoder.fit_transform(df['Rating text'])
df.head()
correlation_matrix = df[['Latitude', 'Longitude', 'Rating text']].corr()
print(correlation_matrix)
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```
