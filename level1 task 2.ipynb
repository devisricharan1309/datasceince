```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("Dataset .csv")
df.describe()
```

```python
df[["Average Cost for two","Price range","Aggregate rating","Votes"]].describe()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Country Code",data=df,palette="cividis")
plt.title("DISTRIBUTION OF RESTAURANTS BY COUNTRY CODE")
plt.xlabel("Country Code")
plt.ylabel("No.of restaurants")
plt.show()
```

```python
top_countries=df["Country Code"].value_counts().head()
print("Top 5 countries with highest no.of restaurants")
print(top_countries)
```

```python
plt.figure(figsize=(15,6))
sns.countplot(x="City",data=df,order=df["City"].value_counts().head(20).index)
plt.title("DISTRIBUTION OF RESTAURANTS BY CITY")
plt.xlabel("City")
plt.ylabel("No.of Restaurants")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(15,6))
cuisines_count=df["Cuisines"].value_counts()
cuisines_count.head(25).plot(kind="bar",color=sns.color_palette("Set3",15))
plt.title("TOP 20 CUISINES WITH HIGHEST NUMBER OF RESTAURANTS")
plt.xlabel("Cuisine")
plt.ylabel("No.of Restaurants")
plt.xticks(rotation=45)
plt.show()
```

```python
top_cities=df["City"].value_counts().head(10)
print("Top 10 cities with highest no.of restaurants")
print(top_cities)
```

```python
top_cuisines=cuisines_count.head(10)
print("Top 10 cuisines with highest no.of restaurants")
print(top_cuisines)
```
